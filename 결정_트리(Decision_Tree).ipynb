# 결정 트리(Decision Tree)

* 분류와 회귀에 사용되는 지도 학습 방법
* 데이터 특성으로 부터 추론된 결정 규칙을 통해 값을 예측
* **if-then-else** 결정 규칙을 통해 데이터 학습
* 트리의 깊이가 깊을 수록 복잡한 모델
* 결정 트리 장점
  * 이해와 해석이 쉽다
  * 시각화가 용이하다
  * 많은 데이터 전처리가 필요하지 않다
  * 수치형과 범주형 데이터 모두를 다룰 수 있다
  * ...

## 분류 - `DecisionTreeClassifier()`

* `DecisionTreeClassifier`는 분류를 위한 결정트리 모델
* 두개의 배열 X, y를 입력 받음
  * X는 [n_samples, n_features] 크기의 데이터 특성 배열
  * y는 [n_samples] 크기의 정답 배열

```
X = [[0,0],[1,1]]
y = [0,1]

model = tree.DecisionTreeClassifier()
model.fit(X, y)
```

```
model.predict([[2.,2.]])
```

##### 전처리 없이 학습

```
model = DecisionTreeClassifier()
```

```
cross_val_score(
    estimator = model,
    X = iris.data,
    y = iris.target,
    cv=5,
    n_jobs=multiprocessing.cpu_count()
)
```

##### 전처리 후 학습
* 결정 트리는 규칙을 학습하기 때문에 전처리에 큰 영향을 받지 않는다.

```
model = make_pipeline(
    StandardScaler(),
    DecisionTreeClassifier()
)
```

```
cross_val_score(
    estimator = model,
    X = iris.data,
    y = iris.target,
    cv=5,
    n_jobs=multiprocessing.cpu_count()
)
```

#### 학습된 결정 트리 시각화

```
model = DecisionTreeClassifier()
model.fit(iris.data, iris.target)
```

##### 텍스트를 통한 시각화

```
r = tree.export_text(decision_tree=model, feature_names=iris.feature_names)
print(r)
```

##### plot_tree를 사용한 시각화

```
tree.plot_tree(model)
```

##### graphviz를 사용한 시각화

```
dot_data = tree.export_graphviz(
    decision_tree=model,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True, rounded=True,
    special_characters=True)
graph = graphviz.Source(dot_data)
graph
```

#### 시각화

```
n_classes = 3
plot_colors = "ryb"
plot_step = 0.02
```

##### 결정 경계 시각화

```
plt.figure(figsize=(16,8))

for pairidx, pair in enumerate([[0,1],[0,2],[0,3],[1,2],[1,3],[2,3]]):

    X = iris.data[:,pair]
    y = iris.target

    model = DecisionTreeClassifier()
    model.fit(X, y)

    plt.subplot(2,3,pairidx+1)

    x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
    y_min, y_max = X[:,1].min() - 1, X[:,1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step), np.arange(y_min, y_max, plot_step))

    plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)


    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu)

    plt.xlabel(iris.feature_names[pair[0]])
    plt.ylabel(iris.feature_names[pair[1]])

    for i, color in zip(range(n_classes), plot_colors):
        idx = np.where(y == i)
        plt.scatter(X[idx,0], X[idx,1], c=color, label=iris.target_names[i],
                    cmap=plt.cm.RdYlBu, edgecolor='b', s=15)

plt.suptitle("Decision surface of a decision tree using paired features")
plt.legend(loc='lower right', borderpad=0, handletextpad=0)
plt.axis("tight")
plt.show()
```

##### 하이퍼파리미터를 변경해 보면서 결정 경계의 변화 확인

```
plt.figure(figsize=(16,8))

for pairidx, pair in enumerate([[0,1],[0,2],[0,3],[1,2],[1,3],[2,3]]):

    X = iris.data[:,pair]
    y = iris.target

    model = DecisionTreeClassifier(max_depth=2)
    model.fit(X, y)

    plt.subplot(2,3,pairidx+1)

    x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
    y_min, y_max = X[:,1].min() - 1, X[:,1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step), np.arange(y_min, y_max, plot_step))

    plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)


    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu)

    plt.xlabel(iris.feature_names[pair[0]])
    plt.ylabel(iris.feature_names[pair[1]])

    for i, color in zip(range(n_classes), plot_colors):
        idx = np.where(y == i)
        plt.scatter(X[idx,0], X[idx,1], c=color, label=iris.target_names[i],
                    cmap=plt.cm.RdYlBu, edgecolor='b', s=15)

plt.suptitle("Decision surface of a decision tree using paired features")
plt.legend(loc='lower right', borderpad=0, handletextpad=0)
plt.axis("tight")
plt.show()
```

#### 시각화

##### 회귀식 시각화

```
plt.figure(figsize=(16, 8))

# 사용할 특성의 인덱스 (0, 1, 2 번 특성)
for pairidx in range(3):  # range(3)을 사용하여 인덱스를 생성합니다.
    X = boston.data.iloc[:, pairidx].values.reshape(-1, 1)  # DataFrame에서 특성 선택 및 reshape
    y = boston.target.values  # 타겟값 설정 (Series에서 Numpy 배열로 변환)

    model = DecisionTreeRegressor()  # 회귀 모델 생성
    model.fit(X, y)  # 모델 학습

    # 테스트 데이터 생성
    X_test = np.arange(min(X), max(X), 0.1)[:, np.newaxis]  # 입력 특성의 범위에서 0.1 간격으로 데이터 생성
    predict = model.predict(X_test)  # 예측 수행

    # 서브플롯 설정
    plt.subplot(1, 3, pairidx + 1)  # pairidx + 1 사용
    plt.scatter(X, y, s=20, edgecolor="k", c="darkorange", label="data")  # 데이터 포인트
    plt.plot(X_test, predict, color="royalblue", linewidth=2)  # 예측 결과
    plt.xlabel(boston.feature_names[pairidx])  # x축 레이블
    plt.ylabel('target')  # y축 레이블

plt.tight_layout()  # 서브플롯 간의 간격 조정
plt.show()  # 플롯 표시
```

##### 하이퍼파리미터를 변경해 보면서 회귀식 시각화

```
plt.figure(figsize=(16, 8))

# 사용할 특성의 인덱스 (0, 1, 2 번 특성)
for pairidx in range(3):  # range(3)을 사용하여 인덱스를 생성합니다.
    X = boston.data.iloc[:, pairidx].values.reshape(-1, 1)  # DataFrame에서 특성 선택 및 reshape
    y = boston.target.values  # 타겟값 설정 (Series에서 Numpy 배열로 변환)

    model = DecisionTreeRegressor(max_depth=3)  # 회귀 모델 생성
    model.fit(X, y)  # 모델 학습

    # 테스트 데이터 생성
    X_test = np.arange(min(X), max(X), 0.1)[:, np.newaxis]  # 입력 특성의 범위에서 0.1 간격으로 데이터 생성
    predict = model.predict(X_test)  # 예측 수행

    # 서브플롯 설정
    plt.subplot(1, 3, pairidx + 1)  # pairidx + 1 사용
    plt.scatter(X, y, s=20, edgecolor="k", c="darkorange", label="data")  # 데이터 포인트
    plt.plot(X_test, predict, color="royalblue", linewidth=2)  # 예측 결과
    plt.xlabel(boston.feature_names[pairidx])  # x축 레이블
    plt.ylabel('target')  # y축 레이블

plt.tight_layout()  # 서브플롯 간의 간격 조정
plt.show()
```
